1.1 Inputs

Quiz 1 mark
Quiz 2 mark
Quiz 3 mark
Continue choice (Y/N)

1.2 Process

Read the three quiz marks.
Calculate the average.
Compare the average with 50.
Ask whether to continue.

1.3 Output

Average mark.
PASS or FAIL.
Program Ended.

Pseudocode

START
SET choice = "Y"

WHILE choice = "Y"
    INPUT quiz_1
    INPUT quiz_2
    INPUT quiz_3

    average = (quiz_1 + quiz_2 + quiz_3) / 3

    DISPLAY average

    IF average >= 50
        DISPLAY "PASS"
    ELSE
        DISPLAY "FAIL"
    ENDIF

    INPUT choice
ENDWHILE

DISPLAY "Program Ended"
END

Algorithm Structure

Sequence: Input marks → Calculate average → Display average.
Selection: IF average >= 50 THEN PASS ELSE FAIL.
Iteration: WHILE choice == "Y".
